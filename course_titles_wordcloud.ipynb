```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 1. Base list of the actual course titles from your project data
course_titles = [
    'python for data science', 'introduction to data science', 'big data 101', 
    'hadoop 101', 'data analysis with python', 'data science methodology', 
    'machine learning with python', 'spark fundamentals i', 
    'data science hands on with open source tools', 'blockchain essentials', 
    'data visualization with python', 'deep learning 101', 'build your own chatbot', 
    'r for data science', 'statistics 101', 'introduction to cloud', 
    'docker essentials a developer introduction', 'sql and relational databases 101', 
    'mapreduce and yarn', 'data privacy fundamentals'
]

# 2. Combine all titles into a single text string
# (Optional: Weights can be simulated by repeating titles based on their relative enrollment rankings)
text_corpus = " ".join(course_titles)

# 3. Initialize and configure the WordCloud object
wordcloud = WordCloud(
    width=800, 
    height=450, 
    background_color='white',   # Clear presentation background
    colormap='tab10',           # Clean, distinct professional color palette
    max_words=150,              # Limits visual clutter
    collocations=False          # Prevents duplicate keyword phrases from sticking together
).generate(text_corpus)

# 4. Plot the generated word cloud using matplotlib
plt.figure(figsize=(10, 6), dpi=300)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide grid axes lines for presentation clarity
plt.tight_layout(pad=0)

# 5. Export the clean high-resolution image file for your presentation slide
plt.savefig('course_titles_wordcloud.png', dpi=300, bbox_inches='tight')
print("Visual Asset saved as: course_titles_wordcloud.png")
```
