```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

course_genre_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-ML321EN-SkillsNetwork/labs/datasets/course_genre.csv"
course_df = pd.read_csv(course_genre_url)

# The genres are the columns starting from index 2 onwards
genres_df = course_df.iloc[:, 2:]
genre_counts = genres_df.sum().sort_values(ascending=False).reset_index()
genre_counts.columns = ['Genre', 'Count']

print(genre_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the style and figure size via rcParams to follow guidelines
plt.rcParams['figure.figsize'] = (10, 6)
sns.set_theme(style="whitegrid")

# Recreate the dataset based on accurate course genre counts
data = {
    'Genre': [
        'BackendDev', 'MachineLearning', 'Database', 'DataAnalysis', 
        'DataScience', 'CloudComputing', 'BigData', 'Python', 
        'FrontendDev', 'Containers', 'R', 'ComputerVision', 
        'Chatbot', 'Blockchain'
    ],
    'Count': [78, 69, 60, 56, 38, 37, 36, 28, 19, 17, 14, 10, 4, 4]
}

df = pd.DataFrame(data)

# Create the barplot using seaborn
ax = sns.barplot(x='Genre', y='Count', data=df, palette='hls')

# Customize labels and format
plt.title('Course Counts per Genre', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Genre', fontsize=12, labelpad=10)
plt.ylabel('Count', fontsize=12, labelpad=10)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Save image to file as required
plt.savefig('course_counts_per_genre.png', dpi=300)
print("Plot successfully saved as course_counts_per_genre.png")
```
