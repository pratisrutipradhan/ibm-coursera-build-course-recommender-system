```python
import pandas as pd
import matplotlib.pyplot as plt

# Recreate the data matching image_1a1843.png
data = {
    'TITLE': [
        'python for data science', 'introduction to data science', 'big data 101', 
        'hadoop 101', 'data analysis with python', 'data science methodology', 
        'machine learning with python', 'spark fundamentals i', 
        'data science hands on with open source tools', 'blockchain essentials', 
        'data visualization with python', 'deep learning 101', 'build your own chatbot', 
        'r for data science', 'statistics 101', 'introduction to cloud', 
        'docker essentials a developer introduction', 'sql and relational databases 101', 
        'mapreduce and yarn', 'data privacy fundamentals'
    ],
    'Enrolls': [
        14936, 14477, 13291, 10599, 8303, 7719, 7644, 7551, 7199, 6719, 
        6709, 6323, 5512, 5237, 5015, 4983, 4480, 3697, 3670, 3624
    ]
}

df = pd.DataFrame(data)

# Create a styled dataframe table image to exactly replicate the presentation style in image_1a1843.png
fig, ax = plt.subplots(figsize=(6.5, 7.5), dpi=300)
ax.axis('off')

# Custom table drawing to look like the styled dataframe render in image_1a1843.png
table = ax.table(
    cellText=df.values,
    rowLabels=df.index,
    colLabels=df.columns,
    loc='center',
    cellLoc='right',
    colWidths=[0.75, 0.18]
)

# Align the title column text to the right as in the pandas display style seen in the screenshot
for (row, col), cell in table.get_celld().items():
    if col == 0 and row > 0:
        cell.set_text_props(ha='right')
    # Style fonts and borders to match classic dataframe clean display
    cell.set_edgecolor('#E0E0E0')
    if row == 0:
        cell.set_text_props(fontweight='bold', color='#333333')
        cell.set_edgecolor('white')
    if col == -1:
        cell.set_text_props(fontweight='bold')
        cell.set_edgecolor('white')

plt.tight_layout()
plt.savefig('popular_courses_dataframe.png', bbox_inches='tight', dpi=300)
print("Dataframe table visualization generated successfully.")
```
