```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size and theme styling
plt.rcParams['figure.figsize'] = (7, 4.5)
plt.rcParams['figure.dpi'] = 300
sns.set_theme(style="ticks")

# Simulating the exact structural distribution pattern observed in the sample image
np.random.seed(42)

# Replicating the distinct peaks: a massive spike at 1, a gap/drop, and a secondary high peak around 5 rolling into a long tail
group_1 = np.repeat(1, 8300)
group_2 = np.repeat(2, 400)
group_3 = np.repeat(3, 100)
group_5_plus = np.random.negative_binomial(n=3, p=0.25, size=28000) + 5
# Clip the tail to match the maximum range around 60 seen in the sample
group_5_plus = group_5_plus[group_5_plus <= 63]

# Combine into a single series
all_enrolls = np.concatenate([group_1, group_2, group_3, group_5_plus])
df_hist = pd.DataFrame({'enrolls': all_enrolls})

# Create the histogram matching the visual style of the screenshot
ax = sns.histplot(data=df_hist, x='enrolls', bins=100, edgecolor='white', linewidth=0.5, color='#4682B4')

# Labeling and Formatting matching the sample exact axis structure
plt.xlabel('enrolls', fontsize=11)
plt.ylabel('Count', fontsize=11)
plt.xlim(-2, 65)
plt.ylim(0, 8900)

# Clean adjustments
plt.tight_layout()

# Save image
plt.savefig('course_enrollment_histogram.png', dpi=300)
print("Histogram successfully saved as course_enrollment_histogram.png")
```
